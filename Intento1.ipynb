```python
#Adivina y verifica enumeracion, debe tener resultado exacto
objetivo = int(input(" escoge un entero"))
respuesta = 0

while respuesta **2 < objetivo: #podemos usar print para ver que hacen las cosas que no entendemos
    print (respuesta)
    respuesta += 1

if respuesta**2 == objetivo:
  print(f"la raiz cuadrada de {objetivo} es {respuesta}")

else:
  print(f"{objetivo} no tiene raiz cuadrada exacta")

```

```python
#aproximacion soluciones
#se dbee definir que tan cerca queremos estar de la solucion, no necesita
#respuesta exacta
#epsilon que tan cerca queremos estar de esa solucion
#trade canje o ser rapidos o precisos
#mas ciclos de computo para ser mas precisos

#ejericio raiz cuadrada

objetivo = int(input("escoge un numero "))
epsilon = 0.01 #que tan precisos queremos ser
paso = epsilon ** 2 #que tanto vamos a acercarnos a cada iteracion
respuesta = 0.0

#abs nos regresa el valor absoluto

while abs(respuesta ** 2 - objetivo) >= epsilon and respuesta <= objetivo: #para evitar numeros negativos
    print (respuesta ** 2)
    respuesta += paso

if abs(respuesta **2 - objetivo) >= epsilon:
    print (f'no se encontro la raiz cuadrada de {objetivo}')

else:
    print (f'la raiz cuadrada de {objetivo} es {respuesta}')

```

```python
#busqueda binaria requisito conjunto de busqueda debe estar organizado
#con busqueda de raices los numeros ya estan ordenados

objetivo = int(input("Escoge un numero: "))
epsilon = 0.01
bajo = 0.0 #nuestro limite inferior
alto = max(1,0, objetivo) #limite maximo
respuesta = (alto + bajo) /2

while abs (respuesta**2 - objetivo) >= epsilon:
    print(f"bajo={bajo}, alto={alto}, respuesta={respuesta}")
    if respuesta**2 < objetivo:
      bajo = respuesta
    else:
      alto = respuesta

    respuesta = (alto + bajo) / 2

print (f"la raiz cuadrada de {objetivo} es {respuesta}")

```
